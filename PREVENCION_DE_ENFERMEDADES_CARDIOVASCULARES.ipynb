```python
PATH = 'gdrive/MyDrive/DIPLOMADOS/UNAP/IA/'
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
import subprocess
import sys

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

required_packages = [
    'pandas', 'numpy', 'matplotlib', 'seaborn',
    'scikit-learn', 'xgboost', 'catboost',
    'tensorflow', 'keras', 'shap', 'lime', 'imbalanced-learn', 'openpyxl'
]

print("Instalando paquetes necesarios...")
for pkg in required_packages:
    try:
        __import__(pkg.replace('-', '_'))
    except ImportError:
        print(f"Instalando {pkg}...")
        install(pkg)

print("¡Dependencias listas!\n")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier
from catboost import CatBoostClassifier

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import shap
from imblearn.over_sampling import SMOTE
import pickle

# =============================================
# PASO 1: CARGAR Y RENOMBRAR COLUMNAS
# =============================================
file_path = '/content/' + PATH + 'dataset.xlsx'
print(f"Cargando dataset desde: {file_path}")

df = pd.read_excel(file_path)

# Renombrar columnas al español (para coherencia con tu proyecto)
df.rename(columns={
    'male': 'sexo',
    'age': 'edad',
    'education': 'educacion',
    'currentSmoker': 'fuma_actualmente',
    'cigsPerDay': 'cigarrillos_dia',
    'BPMeds': 'medicacion_presion',
    'prevalentStroke': 'ha_tenido_acv',
    'prevalentHyp': 'hipertension_diagnosticada',
    'diabetes': 'diabetes',
    'totChol': 'colesterol_total',
    'sysBP': 'presion_sistolica',
    'diaBP': 'presion_diastolica',
    'BMI': 'imc',
    'heartRate': 'frecuencia_cardiaca',
    'glucose': 'glucosa',
    'TenYearCHD': 'riesgo_coronario_10_años'
}, inplace=True)

print("Primeras 5 filas (columnas renombradas):")
display(df.head())

print("\nTipos de datos:")
print(df.dtypes)

print("\nResumen estadístico:")
display(df.describe())

# =============================================
# PASO 2: LIMPIEZA DE DATOS
# =============================================
print(f"\nForma original: {df.shape}")
print("Valores nulos por columna:")
print(df.isnull().sum())

# Eliminar filas con valores nulos
df = df.dropna()
df = df.drop_duplicates()
print(f"Forma después de limpieza: {df.shape}")

# Asegurar tipo entero para la variable objetivo
df['riesgo_coronario_10_años'] = df['riesgo_coronario_10_años'].astype(int)

# =============================================
# PASO 3: ANÁLISIS EXPLORATORIO
# =============================================
# Distribución de clases
plt.figure(figsize=(7,5))
sns.countplot(x='riesgo_coronario_10_años', data=df, palette='Set1')
plt.title('Distribución de Riesgo Coronario a 10 años')
plt.xlabel('Riesgo (0 = No, 1 = Sí)')
plt.ylabel('Frecuencia')
plt.xticks([0, 1], ['No', 'Sí'])
plt.show()

print("\nDistribución de clases (%):")
print(df['riesgo_coronario_10_años'].value_counts(normalize=True) * 100)

# Matriz de correlación
plt.figure(figsize=(14,10))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.tight_layout()
plt.show()

# Histogramas
df.hist(figsize=(16,12), bins=20, color='lightblue', edgecolor='black')
plt.suptitle('Distribución de Variables Numéricas', fontsize=16)
plt.tight_layout()
plt.show()

# =============================================
# PASO 4: BALANCEO CON SMOTE
# =============================================
X = df.drop('riesgo_coronario_10_años', axis=1)
y = df['riesgo_coronario_10_años']

min_class_ratio = y.value_counts(normalize=True).min()
print(f"\nProporción clase minoritaria: {min_class_ratio:.3f}")

if min_class_ratio < 0.4:
    print("Desbalance detectado → Aplicando SMOTE...")
    smote = SMOTE(random_state=42)
    X_res, y_res = smote.fit_resample(X, y)
    print(f"Nuevo tamaño: {X_res.shape[0]} muestras")
else:
    print("Clases balanceadas → No se aplica SMOTE.")
    X_res, y_res = X, y

# =============================================
# PASO 5: ESCALADO Y DIVISIÓN
# =============================================
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_res), columns=X_res.columns)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_res, test_size=0.3, random_state=42, stratify=y_res
)

print(f"\nEntrenamiento: {X_train.shape}, Prueba: {X_test.shape}")

# =============================================
# PASO 6: MODELOS
# =============================================
models = {}
predictions = {}
probabilities = {}

# --- Random Forest ---
rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
predictions['Random Forest'] = rf.predict(X_test)
probabilities['Random Forest'] = rf.predict_proba(X_test)[:, 1]
models['Random Forest'] = rf

# --- XGBoost ---
xgb = XGBClassifier(random_state=42, eval_metric='logloss', n_estimators=200)
xgb.fit(X_train, y_train)
predictions['XGBoost'] = xgb.predict(X_test)
probabilities['XGBoost'] = xgb.predict_proba(X_test)[:, 1]
models['XGBoost'] = xgb

# --- CatBoost ---
cat = CatBoostClassifier(random_state=42, verbose=0, iterations=200)
cat.fit(X_train, y_train)
predictions['CatBoost'] = cat.predict(X_test)
probabilities['CatBoost'] = cat.predict_proba(X_test)[:, 1]
models['CatBoost'] = cat

# --- Red Neuronal (MLP) ---
mlp = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
mlp.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
mlp.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)

mlp_pred_proba = mlp.predict(X_test).flatten()
mlp_pred = (mlp_pred_proba > 0.5).astype(int)
predictions['MLP'] = mlp_pred
probabilities['MLP'] = mlp_pred_proba
models['MLP'] = mlp

# =============================================
# PASO 7: EVALUACIÓN
# =============================================
metrics = {}
for name in models:
    y_pred = predictions[name]
    y_prob = probabilities[name]
    metrics[name] = {
        'Accuracy': round(accuracy_score(y_test, y_pred), 4),
        'Precision': round(precision_score(y_test, y_pred), 4),
        'Recall': round(recall_score(y_test, y_pred), 4),
        'F1-Score': round(f1_score(y_test, y_pred), 4),
        'AUC-ROC': round(roc_auc_score(y_test, y_prob), 4)
    }

metrics_df = pd.DataFrame(metrics).T
print("\n" + "="*70)
print("          TABLA COMPARATIVA DE MODELOS")
print("="*70)
display(metrics_df)

# =============================================
# MEJOR MODELO
# =============================================
best_model_name = metrics_df['F1-Score'].idxmax()
best_model = models[best_model_name]
print(f"\nMEJOR MODELO: {best_model_name}")
print(f"F1-Score: {metrics_df.loc[best_model_name, 'F1-Score']}")
print(f"AUC-ROC: {metrics_df.loc[best_model_name, 'AUC-ROC']}")

# =============================================
# GRÁFICOS
# =============================================
# 1. Matriz de confusión
y_pred_best = predictions[best_model_name]
cm = confusion_matrix(y_test, y_pred_best)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['No Riesgo', 'Riesgo'], yticklabels=['No Riesgo', 'Riesgo'])
plt.title(f'Matriz de Confusión - {best_model_name}')
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.show()

# 2. Curva ROC
plt.figure(figsize=(10,7))
for name in models:
    fpr, tpr, _ = roc_curve(y_test, probabilities[name])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.3f})')
plt.plot([0,1], [0,1], 'k--', label='Aleatorio')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC Comparativa')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 3. Importancia de variables
if best_model_name != 'MLP':
    importances = best_model.feature_importances_
    feat_imp = pd.Series(importances, index=X_train.columns).sort_values(ascending=False)
    plt.figure(figsize=(10,6))
    sns.barplot(x=feat_imp.values, y=feat_imp.index, palette='magma')
    plt.title(f'Importancia de Variables - {best_model_name}')
    plt.xlabel('Importancia Relativa')
    plt.tight_layout()
    plt.show()

# =============================================
# INTERPRETACIÓN CON SHAP
# =============================================
print("\nGenerando gráfico SHAP...")
try:
    if best_model_name in ['Random Forest', 'XGBoost', 'CatBoost']:
        explainer = shap.TreeExplainer(best_model)
        shap_values = explainer.shap_values(X_test)
        shap.summary_plot(shap_values, X_test, plot_type="bar", max_display=10)
    else:
        explainer = shap.DeepExplainer(mlp, X_train.values[:100])
        shap_values = explainer.shap_values(X_test.values[:100])
        shap.summary_plot(shap_values[0], X_test.iloc[:100], plot_type="bar")
except Exception as e:
    print(f"SHAP falló: {e}")

# =============================================
# GUARDAR MEJOR MODELO Y SCALER
# =============================================
model_save_path = '/content/' + PATH + 'mejor_modelo_cardio.pkl'
scaler_save_path = '/content/' + PATH + 'scaler_cardio.pkl'

if best_model_name == 'MLP':
    h5_path = model_save_path.replace('.pkl', '.h5')
    mlp.save(h5_path)
    print(f"\nModelo MLP guardado en: {h5_path}")
else:
    with open(model_save_path, 'wb') as f:
        pickle.dump(best_model, f)
    print(f"\nModelo {best_model_name} guardado en: {model_save_path}")

with open(scaler_save_path, 'wb') as f:
    pickle.dump(scaler, f)
print(f"Scaler guardado en: {scaler_save_path}")

print("\n¡PROYECTO COMPLETADO CON ÉXITO!")
print(f"Archivos guardados en: {PATH}")
```

```python
PATH = 'gdrive/MyDrive/DIPLOMADOS/UNAP/IA/'
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

!pip install pandas numpy matplotlib seaborn scikit-learn xgboost catboost tensorflow keras shap lime imbalanced-learn openpyxl

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, precision_recall_curve
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from catboost import CatBoostClassifier

from imblearn.combine import SMOTETomek
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
import pickle

# =============================================
# PASO 1: CARGAR Y RENOMBRAR COLUMNAS
# =============================================
file_path = '/content/' + PATH + 'dataset.xlsx'
print(f"Cargando dataset desde: {file_path}")
df = pd.read_excel(file_path)

df.rename(columns={
    'male': 'sexo',
    'age': 'edad',
    'education': 'educacion',
    'currentSmoker': 'fuma_actualmente',
    'cigsPerDay': 'cigarrillos_dia',
    'BPMeds': 'medicacion_presion',
    'prevalentStroke': 'ha_tenido_acv',
    'prevalentHyp': 'hipertension_diagnosticada',
    'diabetes': 'diabetes',
    'totChol': 'colesterol_total',
    'sysBP': 'presion_sistolica',
    'diaBP': 'presion_diastolica',
    'BMI': 'imc',
    'heartRate': 'frecuencia_cardiaca',
    'glucose': 'glucosa',
    'TenYearCHD': 'riesgo_coronario_10_años'
}, inplace=True)

df = df.dropna().drop_duplicates()
df['riesgo_coronario_10_años'] = df['riesgo_coronario_10_años'].astype(int)
print(f"Dataset limpio: {df.shape}")

# =============================================
# PASO 2: FEATURE ENGINEERING
# =============================================
df['presion_media'] = (df['presion_sistolica'] + 2*df['presion_diastolica']) / 3
df['colesterol_imc'] = df['colesterol_total'] / (df['imc'] + 1)
df['edad_glucosa'] = df['edad'] * df['glucosa']
df['riesgo_hipertension'] = df['presion_sistolica'] / (df['edad'] + 1)

# =============================================
# PASO 3: BALANCEO DE CLASES (SMOTETomek)
# =============================================
X = df.drop('riesgo_coronario_10_años', axis=1)
y = df['riesgo_coronario_10_años']

smote_tomek = SMOTETomek(random_state=42)
X_res, y_res = smote_tomek.fit_resample(X, y)
print("Balanceo aplicado →", X_res.shape)

# =============================================
# PASO 4: ESCALADO Y DIVISIÓN
# =============================================
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_res), columns=X_res.columns)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_res, test_size=0.3, random_state=42, stratify=y_res
)
print("Tamaño entrenamiento:", X_train.shape)

# =============================================
# PASO 5: OPTIMIZACIÓN RANDOM FOREST
# =============================================
param_grid = {
    'n_estimators': [200, 500, 800],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

rf_random = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    param_distributions=param_grid,
    n_iter=15,
    scoring='precision',
    cv=5,
    n_jobs=-1,
    random_state=42
)
rf_random.fit(X_train, y_train)
rf_best = rf_random.best_estimator_
print("Mejores parámetros RF:", rf_random.best_params_)

# =============================================
# PASO 6: ENTRENAR OTROS MODELOS
# =============================================
xgb = XGBClassifier(random_state=42, eval_metric='logloss', n_estimators=300, learning_rate=0.05)
xgb.fit(X_train, y_train)

cat = CatBoostClassifier(random_state=42, verbose=0, iterations=400, depth=8, learning_rate=0.05)
cat.fit(X_train, y_train)

# Red Neuronal
mlp = Sequential([
    Dense(256, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
mlp.compile(optimizer='adam', loss='binary_crossentropy', metrics=['Precision'])
early_stop = EarlyStopping(monitor='val_precision', mode='max', patience=20, restore_best_weights=True)
mlp.fit(X_train, y_train, epochs=200, batch_size=32, validation_split=0.2, callbacks=[early_stop], verbose=0)

# =============================================
# PASO 7: AJUSTE DE UMBRAL PARA PRECISIÓN
# =============================================
def best_threshold_for_precision(model, X_test, y_test):
    probs = model.predict_proba(X_test)[:, 1]
    precisions, recalls, thresholds = precision_recall_curve(y_test, probs)
    opt_thresh = thresholds[np.argmax(precisions)]
    opt_precision = max(precisions)
    return opt_thresh, opt_precision

models = {
    'Random Forest': rf_best,
    'XGBoost': xgb,
    'CatBoost': cat
}

metrics = {}
for name, model in models.items():
    th, best_prec = best_threshold_for_precision(model, X_test, y_test)
    preds = (model.predict_proba(X_test)[:, 1] >= th).astype(int)
    metrics[name] = {
        'Accuracy': round(accuracy_score(y_test, preds), 4),
        'Precision': round(precision_score(y_test, preds), 4),
        'Recall': round(recall_score(y_test, preds), 4),
        'F1-Score': round(f1_score(y_test, preds), 4),
        'AUC-ROC': round(roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]), 4),
        'Umbral Óptimo': round(th, 3)
    }

# =============================================
# PASO 8: ENSAMBLE FINAL (STACKING)
# =============================================
stack_model = StackingClassifier(
    estimators=[
        ('rf', rf_best),
        ('xgb', xgb),
        ('cat', cat)
    ],
    final_estimator=LogisticRegression(),
    n_jobs=-1
)
stack_model.fit(X_train, y_train)

stack_probs = stack_model.predict_proba(X_test)[:, 1]
stack_preds = (stack_probs >= 0.6).astype(int)
metrics['Stacking'] = {
    'Accuracy': round(accuracy_score(y_test, stack_preds), 4),
    'Precision': round(precision_score(y_test, stack_preds), 4),
    'Recall': round(recall_score(y_test, stack_preds), 4),
    'F1-Score': round(f1_score(y_test, stack_preds), 4),
    'AUC-ROC': round(roc_auc_score(y_test, stack_probs), 4),
    'Umbral Óptimo': 0.6
}

# =============================================
# PASO 9: RESULTADOS
# =============================================
metrics_df = pd.DataFrame(metrics).T.sort_values('Precision', ascending=False)
print("\n==================== RESULTADOS FINALES ====================")
display(metrics_df)

# =============================================
# PASO 10: GUARDAR MODELOS
# =============================================
best_name = metrics_df.index[0]
best_model = models.get(best_name, stack_model)

with open(PATH + 'mejor_modelo_preciso.pkl', 'wb') as f:
    pickle.dump(best_model, f)
with open(PATH + 'scaler_preciso.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print(f"\n✅ Mejor modelo: {best_name}")
print(f"Precisión: {metrics_df.loc[best_name, 'Precision']}")
print(f"Guardado en: {PATH}")
```

```python
# =============================================
# CURVA ROC COMPARATIVA DE MODELOS
# =============================================

from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

# Configuración de estilo para mejores gráficos
plt.style.use('default')
sns.set_palette("husl")

# =============================================
# CALCULAR CURVAS ROC PARA CADA MODELO
# =============================================

# Diccionario para almacenar las curvas ROC
roc_data = {}

# 1. Random Forest
rf_probs = rf_best.predict_proba(X_test)[:, 1]
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_probs)
roc_auc_rf = auc(fpr_rf, tpr_rf)
roc_data['Random Forest'] = (fpr_rf, tpr_rf, roc_auc_rf)

# 2. XGBoost
xgb_probs = xgb.predict_proba(X_test)[:, 1]
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, xgb_probs)
roc_auc_xgb = auc(fpr_xgb, tpr_xgb)
roc_data['XGBoost'] = (fpr_xgb, tpr_xgb, roc_auc_xgb)

# 3. CatBoost
cat_probs = cat.predict_proba(X_test)[:, 1]
fpr_cat, tpr_cat, _ = roc_curve(y_test, cat_probs)
roc_auc_cat = auc(fpr_cat, tpr_cat)
roc_data['CatBoost'] = (fpr_cat, tpr_cat, roc_auc_cat)

# 4. Red Neuronal (MLP)
mlp_probs = mlp.predict(X_test).flatten()
fpr_mlp, tpr_mlp, _ = roc_curve(y_test, mlp_probs)
roc_auc_mlp = auc(fpr_mlp, tpr_mlp)
roc_data['Red Neuronal'] = (fpr_mlp, tpr_mlp, roc_auc_mlp)

# 5. Stacking
stack_probs = stack_model.predict_proba(X_test)[:, 1]
fpr_stack, tpr_stack, _ = roc_curve(y_test, stack_probs)
roc_auc_stack = auc(fpr_stack, tpr_stack)
roc_data['Stacking'] = (fpr_stack, tpr_stack, roc_auc_stack)

# =============================================
# GRÁFICO PRINCIPAL - CURVA ROC COMPARATIVA
# =============================================

plt.figure(figsize=(12, 8))

# Colores para cada modelo
colors = ['blue', 'green', 'red', 'orange', 'purple', 'brown']

# Plotear cada curva ROC
for i, (model_name, (fpr, tpr, auc_score)) in enumerate(roc_data.items()):
    plt.plot(fpr, tpr, color=colors[i], lw=2.5,
             label=f'{model_name} (AUC = {auc_score:.4f})')

# Línea de referencia (clasificador aleatorio)
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--',
         label='Clasificador Aleatorio (AUC = 0.5000)')

# Personalizar el gráfico
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.01])
plt.xlabel('Tasa de Falsos Positivos (False Positive Rate)', fontsize=12, fontweight='bold')
plt.ylabel('Tasa de Verdaderos Positivos (True Positive Rate)', fontsize=12, fontweight='bold')
plt.title('CURVA ROC COMPARATIVA - MODELOS DE PREDICCIÓN DE RIESGO CORONARIO',
          fontsize=14, fontweight='bold', pad=20)
plt.legend(loc='lower right', fontsize=11)
plt.grid(True, alpha=0.3)

# Añadir cuadrícula secundaria
plt.minorticks_on()
plt.grid(True, which='minor', alpha=0.2)

plt.tight_layout()
plt.show()

# =============================================
# GRÁFICO ZOOM - ÁREA DE MAYOR INTERÉS
# =============================================

plt.figure(figsize=(10, 8))

# Plotear cada curva ROC con zoom en esquina superior izquierda
for i, (model_name, (fpr, tpr, auc_score)) in enumerate(roc_data.items()):
    plt.plot(fpr, tpr, color=colors[i], lw=2.5,
             label=f'{model_name} (AUC = {auc_score:.4f})')

plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')

# Zoom en la esquina superior izquierda (donde suelen estar los mejores modelos)
plt.xlim([-0.01, 0.5])
plt.ylim([0.5, 1.01])
plt.xlabel('Tasa de Falsos Positivos (False Positive Rate)', fontsize=12, fontweight='bold')
plt.ylabel('Tasa de Verdaderos Positivos (True Positive Rate)', fontsize=12, fontweight='bold')
plt.title('CURVA ROC - ZOOM (Esquina Superior Izquierda)', fontsize=14, fontweight='bold', pad=20)
plt.legend(loc='lower right', fontsize=11)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================
# TABLA COMPARATIVA DE AUC-ROC
# =============================================

print("="*70)
print("COMPARATIVA DE AUC-ROC - TODOS LOS MODELOS")
print("="*70)

auc_comparison = []
for model_name, (fpr, tpr, auc_score) in roc_data.items():
    auc_comparison.append({
        'Modelo': model_name,
        'AUC-ROC': f'{auc_score:.4f}',
        'Interpretación': 'Excelente' if auc_score >= 0.9 else
                         'Muy Bueno' if auc_score >= 0.8 else
                         'Bueno' if auc_score >= 0.7 else
                         'Regular' if auc_score >= 0.6 else 'Pobre'
    })

auc_df = pd.DataFrame(auc_comparison).sort_values('AUC-ROC', ascending=False)
display(auc_df)

# =============================================
# ANÁLISIS DEL MEJOR MODELO SEGÚN AUC-ROC
# =============================================

best_auc_model = auc_df.iloc[0]
print(f"\n🏆 MEJOR MODELO POR AUC-ROC: {best_auc_model['Modelo']}")
print(f"   AUC-ROC: {best_auc_model['AUC-ROC']} ({best_auc_model['Interpretación']})")

# =============================================
# GRÁFICO ADICIONAL: DISTRIBUCIÓN DE SCORES
# =============================================

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

for i, (model_name, (fpr, tpr, auc_score)) in enumerate(roc_data.items()):
    if i < 6:  # Asegurar que no excedamos el número de subplots
        probs = {
            'Random Forest': rf_probs,
            'XGBoost': xgb_probs,
            'CatBoost': cat_probs,
            'Red Neuronal': mlp_probs,
            'Stacking': stack_probs
        }[model_name]

        # Histograma de probabilidades por clase
        axes[i].hist(probs[y_test == 0], bins=30, alpha=0.7, label='Sin Riesgo', color='blue')
        axes[i].hist(probs[y_test == 1], bins=30, alpha=0.7, label='Con Riesgo', color='red')
        axes[i].set_title(f'{model_name}\nAUC = {auc_score:.4f}', fontweight='bold')
        axes[i].set_xlabel('Probabilidad Predicha')
        axes[i].set_ylabel('Frecuencia')
        axes[i].legend()
        axes[i].grid(True, alpha=0.3)

# Ocultar el último subplot si no se usa
if len(roc_data) < 6:
    axes[-1].set_visible(False)

plt.suptitle('DISTRIBUCIÓN DE PROBABILIDADES PREDICHAS POR MODELO',
             fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()

# =============================================
# INTERPRETACIÓN DE RESULTADOS AUC-ROC
# =============================================

print("\n" + "="*70)
print("INTERPRETACIÓN DE LAS CURVAS ROC")
print("="*70)
print("• AUC > 0.9: Excelente capacidad discriminativa")
print("• AUC 0.8-0.9: Muy buena capacidad discriminativa")
print("• AUC 0.7-0.8: Buena capacidad discriminativa")
print("• AUC 0.6-0.7: Capacidad discriminativa regular")
print("• AUC 0.5-0.6: Pobre capacidad discriminativa")
print("• AUC = 0.5: No mejor que el azar")
print("\nLa curva ROC muestra el trade-off entre:")
print("  - Sensibilidad (Verdaderos Positivos)")
print("  - Especificidad (1 - Falsos Positivos)")
```
